```python
#relationship betwwen total positive cases & recovered
sn.relplot(x="TotalPositiveCases",y="Recovered",data=data)
#CONCLUSION: people are recovering from the disease as the no. of cases are increasing
```

```python
#relationship betwwen total positive cases & deaths
sn.relplot(x="TotalPositiveCases",y="Deaths",data=data)
#CONCLUSION: no. of deaths have increased as the no. of cases have increased
```

```python
#relationship betwwen total positive cases & total hospitalized patients
sn.relplot(x="TotalPositiveCases",y="TotalHospitalizedPatients", hue="Recovered",data=data)
#CONCLUSION: upto a certain time (around 40,000 cases)the no. of hospitalisations increased as the no. of cases increased
             #However, after that we see that the no. of hospitalisations decreased as the no. of cases increased
             #Also, the recovery rate increased as the no. of cases increased
```

```python
#pairwise relationship of each parameter
sn.pairplot(data)
```

```python
#line plot
sn.relplot(x="HomeConfinement",y="TotalPositiveCases",kind="line",data=data)
#CONCLUSION: cases still increased as the home confinement increased
```

```python
#line plot
sn.relplot(x="HomeConfinement",y="Recovered",kind="line",data=data)
#CONCLUSION: people are recovering more from the disease as the home confinement is increasing
```

```python
#death rate region-wise
sn.catplot(x="Deaths",y="RegionName",data=data)
#CONLCUSION: Lombardia has the maximum deaths
```

```python
#recovery rate region-wise
sn.catplot(x="Recovered",y="RegionName",data=data)
#CONCLUSION: Lombardia has the maxmimum recoveries
```
